```python
port_mapper = {
    53: 'dns',
    67: 'dhcp-server',
    80: 'http',
    123: 'ntp',
    443: 'https',
    465: 'smtp-ssl',
    993: 'imap-ssl',
    1900: 'ssdp',
    2869: 'icslap',
    3080: 'stm-pproc',
    3478: 'stun',
    3702: 'ws-discovery',
    5000: 'upnp-evnt',
    5222: 'xmpp',
    5228: 'android',
    25050: 'dstPort25050',
    49152: 'dstPort49152',
    49153: 'dstPort49153',
    49154: 'dstPort49154'
}
```

```python
port_keys = list(port_mapper.keys())
port_names = list(port_mapper.values())
```

```python
print(port_names)
```

```python
# for (key, name) in zip(port_keys, port_names):
#     idxs = df[df.dstPort == key].index
#     df.drop(idxs, inplace=True)
```

```python
# set all dst port to 0
for index,row in df.iterrows():
    for k in port_names:
        df.at[index, k] = 0
```

```python
df.head()
```

```python
for index,row in df.iterrows():
    
    if row.dstPort in port_mapper:
        df.at[index, port_mapper[row.dstPort]] = 1
    
    if row.dstPort < 1024:
        df.at[index, 'wellKnownDstPort'] = 1
        df.at[index, 'ephemeralDstPort'] = 0
        df.at[index, 'registedDstPort'] = 0
    if row.dstPort >= 49152:
        df.at[index, 'wellKnownDstPort'] = 0
        df.at[index, 'ephemeralDstPort'] = 1
        df.at[index, 'registedDstPort'] = 0
    else:
        df.at[index, 'wellKnownDstPort'] = 0
        df.at[index, 'ephemeralDstPort'] = 0
        df.at[index, 'registedDstPort'] = 1
```

```python
df[df.dstPort == 80].head()
```

```python
for index,row in df.iterrows():
    if row.srcPort == 49152:
        df.at[index, "srcPort49152"] = 1
        df.at[index, "srcPort49153"] = 0
    elif row.srcPort == 49153:
        df.at[index, "srcPort49152"] = 0
        df.at[index, "srcPort49153"] = 1
    else:
        df.at[index, "srcPort49152"] = 0
        df.at[index, "srcPort49153"] = 0
```
