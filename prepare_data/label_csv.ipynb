```python
df = pd.read_csv('../sample_csv_data/flow.csv')
```

```python
df.head()
```

```python
print(df.shape)
cleansers = ['packetCount', 'totalPayloadByte', 'appplicationPayloadByte', 'duration', 'srcPort', 'dstPort']
for c in cleansers:
    df = df[np.isfinite(df[c])]
print(df.shape)
```

```python
for index,row in df.iterrows():
    if row.srcMac in deviceMap:
        df.set_value(index,'iot', deviceMap[row.srcMac]["iot"])
        df.set_value(index,'device', deviceMap[row.srcMac]["name"])
```

```python
df = df.dropna(subset=['iot', 'device'])
```
